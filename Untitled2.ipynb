```python
import yfinance as yf
import matplotlib.pyplot as plt

# Descargar los datos de Tesla
tesla_data = yf.download('TSLA', start='2020-01-01', end='2023-12-31')

# Restablecer el índice para usar la fecha como columna
tesla_data.reset_index(inplace=True)

# Función para graficar
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Tesla Closing Price', linewidth=2)
    plt.title(title, fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Closing Price (USD)', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.show()

# Dibujar el gráfico de las acciones de Tesla
make_graph(tesla_data, "Tesla Stock Closing Prices (2020-2023)")
```
